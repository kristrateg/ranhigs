**1. Создайте массив со значениями функции y = x**2 **для x = [-2,2] с шагом 0.1**

```python
import numpy as np
def y(x):
    return x**2
```

```python
x_array = np.arange (-2, 2, 0.1)
x_array = np.round(x_array, 2)
print (x_array)
```

```python
y_array = y(x_array)
print (y_array)
print (y_array.shape)
```

**2. Создайте массив со значениями функции y = exp(x)  для x = [-2,2] с шагом 0.1**

```python
def y2(x):
    return np.exp(x)
```

```python
x2_array = np.arange (-2, 2, 0.1)
x2_array = np.round(x2_array, 2)
print (x2_array)
```

```python
y2_array = y2(x2_array)
y2_array = np.round(y2_array, 2)
print (y2_array)
print (y2_array.shape)
```
