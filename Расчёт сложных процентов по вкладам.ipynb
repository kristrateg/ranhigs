Анастасия инвестировала 100 долларов под простые 10 %. Другими словами, ежегодно инвестиция должна приносить 10 % инвестированной суммы, т.е. 10 долларов каждый год: прибыль = 0,10 x исходный баланс.

В то же время Екатерина инвестировала 100 долларов под сложные 5 %. Это значит, что прибыль составит 5 % от текущего баланса, включая предыдущую накопленную прибыль: прибыль = 0,05 х текущий баланс. Екатерина зарабатывает 5 % от 100 долларов в первый год, что дает ей 105 долларов. На следующий год она зарабатывает 5 % от 105 долларов, что составляет 5.25 долларов, и т.д.

**Напишите программу, которая вычислит, сколько лет понадобится для того, чтобы сумма баланса Екатерины превысила сумму баланса Анастасии, с отображением значений обоих балансов за каждый год.**

```python
i = 1

while (100 * ((1+0.05*1)**i)) < (100 * (1 + i*0.1)):    
    print ('Баланс Анастасии в', i, 'год:', int((100 * (1 + i*0.1))))
    print ('Баланс Екатерины в', i, 'год:', int((100 * ((1+0.05*1)**i))))
    i += 1
else:
    print ('Екатерина догонит Анастасию за', i, 'лет')
    print ('Итоговый баланс Екатерины —', int((100 * ((1+0.05*1)**i))))
    print ('Итоговый баланс Анастасии —', int((100 * (1 + i*0.1))))
```
